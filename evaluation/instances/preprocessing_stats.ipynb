```python
from pathlib import Path

from tspn_bnb2.schemas import AnnotatedInstance

# go through all json files and load them
path_non_simplified = Path(
    "/home/krupke/Repositories/tspn/tspn/evaluation/instances/instances_socg"
)
instances_non_simplified = {}
for instance_path in path_non_simplified.glob("*.json"):
    print(f"Loading {instance_path.name}")
    with open(instance_path) as f:
        instance = AnnotatedInstance.model_validate_json(f.read())
        instances_non_simplified[instance_path.stem] = instance
```

```python
path_simplified = Path(
    "/home/krupke/Repositories/tspn/tspn/evaluation/instances/instances_socg_simplified"
)
instances_simplified = {}
for instance_path in path_simplified.glob("*.json"):
    print(f"Loading {instance_path.name}")
    with open(instance_path) as f:
        instance = AnnotatedInstance.model_validate_json(f.read())
        instances_simplified[instance_path.stem] = instance
```

```python
def compare_instances(inst_non_simp: AnnotatedInstance, inst_simp: AnnotatedInstance):
    num_polys = len(inst_non_simp.polygons)
    num_polys_simp = len(inst_simp.polygons)
    num_holes = sum(len(poly.interiors) for poly in inst_non_simp.polygons)
    num_holes_simp = sum(len(poly.interiors) for poly in inst_simp.polygons)
    num_holes_removed = num_holes - num_holes_simp
    print(f"Number of holes: {num_holes} -> {num_holes_simp} (removed {num_holes_removed})")
    num_polys_removed = num_polys - num_polys_simp
    print(f"Number of polygons: {num_polys} -> {num_polys_simp} (removed {num_polys_removed})")


for instance_name in instances_non_simplified:
    print(f"Comparing instance {instance_name}")
    inst_non_simp = instances_non_simplified[instance_name]
    inst_simp = instances_simplified[instance_name]
    compare_instances(inst_non_simp, inst_simp)
```
